```python
%%sql
SELECT COUNT(*) AS TotalCounts FROM movies m
LEFT JOIN genremap gm ON m.movie_id = gm.movie_id
LEFT JOIN genres g ON gm.genre_id = g.genre_id
LEFT JOIN keywordmap keymap ON m.movie_id = keymap.movie_id
LEFT JOIN keywords ON keywords.keyword_id = keymap.keyword_id
WHERE genre_name = "Thriller" AND TRIM(keyword_name) LIKE "%love%"
```
